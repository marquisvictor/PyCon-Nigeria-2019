### Downloading Tweets Dataset. 

Twitter basically has two APIs for developers who have accounts with them to retrieve data, 
1. the Representative State Transfer (REST) API and 
2. the Streaming API

The REST API also known as the Representative state transfer API allows you mine backdated tweets. The Streaming API, however, allows you to gather tweets in real time. One can then infer that they both work differently. The only draw back or constraint with Twitter’s official API is that you can’t get Tweets older than a week (7 days), with rate limits to it, except you are using the Streaming API and can wait for dyas to mine enough big data. More often than not, you realize that you don’t have that time. 

Nonetheless, there are ways we can access and mine older tweets by just scraping the twitter search engine, here, we don’t need Twitter’s Authentication; so, we won’t be affected by Twitter’s rate limits.

### Loading in the Downloaded dataset and Preprocessing after downloading 

````python
import pandas as pd        # Python's library for dataframe manipulation
from pprint import pprint  # Library for neat printing 
import re                  # Python's library for regular expressions 
import matplotlib.pyplot as plt

traffic = pd.read_csv('traffic.csv')# parse_dates=True)     # Loading the dataset to a pandas dataframe 
````

````python
traffic.head(5) # VIewing the first 5 rows in the dataframe
````

````python
traffic = traffic.drop(['geo', 'id', 'permalink'], axis=1) # Dropping irrelevant columns 
traffic.head().transpose()
````

````python
def clean_tweet(tweet):    
    '''
        Utility function to clean tweet text by removing links, special characters
        using simple regex statements.      ''' 
    
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)*_#:.?", " ", tweet).split())

def clean_hashtags(hashtag):    
    '''
        Utility function to clean tweet hashtags by removing the hastags and special characters
        using simple regex statements.
        '''     
    return ' '.join(re.sub("#", " ", hashtag).split())
````

### Applying the custom functions to clean the Tweet text and Hashtag column in the dataframe 

````python
traffic.text = traffic.text.astype(str)    # First change the data type to string.   
traffic.hashtags = traffic.hashtags.astype(str)  # First change the data type to string.
````

````python
traffic['tweet_clean'] = traffic.text.apply(clean_tweet)  # Applying the cleaning function here
traffic['hashtag_clean'] = traffic.hashtags.apply(clean_hashtags)    
````

````python
traffic.head().T
````

At this stage, we have the dataset cleaned and preprocessed in a dataframe already for for any kind of text analytics pipeline we want to run it through, be it time-series analyis, Sentiment Analysis, Topic modelling, Social network Analysis, Visualisations et cetera. But for the purpose of this workshop, we would be creating a sentiment analysis pipeline, Using [VADER Setiment analyser](https://github.com/cjhutto/vaderSentiment), and [TextBlob](https://textblob.readthedocs.io/en/dev/)

### Using TextBlob

#### [follow this link for a more comprehensive tutorial on the Textblob library for natural language processing operations](https://www.analyticsvidhya.com/blog/2018/02/natural-language-processing-for-beginners-using-textblob/)

````python
from textblob import TextBlob
````

````python
def get_tweet_sentiment(tweet):
        '''
        Utility function to classify sentiment of passed tweet
        using textblob's sentiment method
        '''
        # create TextBlob object of tweet text
        analysis = TextBlob(clean_tweet(tweet))
    
        # set sentiment
        if analysis.sentiment.polarity > 0:
            return 'positive'        
        elif analysis.sentiment.polarity == 0:
            return 'neutral'
        
        else:
            return 'negative'
````

### Getting the sentiments polarity for each tweets in the dataset using TextBlob library

````python
traffic['tweet_sentiment'] = traffic.tweet_clean.apply(get_tweet_sentiment)
traffic['hashtag_sentiment'] = traffic.hashtag_clean.apply(get_tweet_sentiment)
````

### Using VADER 

#### [Follow this link for a more comprehensive tutorial on VADER Sentiments](https://medium.com/analytics-vidhya/simplifying-social-media-sentiment-analysis-using-vader-in-python-f9e6ec6fc52f)

````python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()   # Initialize the sentiment intensity analyzer
````

````python
nega = []
posi = []
neut = []
comp = []

conv = list((traffic['tweet_clean']))

for n in conv:
    nega.append(analyzer.polarity_scores(n)['neg'])
    
for p in conv:
    posi.append(analyzer.polarity_scores(p)['pos'])
                
for nt in conv:
    neut.append(analyzer.polarity_scores(nt)['neu'])
                
for c in conv:
    comp.append(analyzer.polarity_scores(c)['compound'])

traffic['Vad_pos_tweets'] = posi
traffic['Vad_neg_tweets'] = nega
traffic['Vad_neutr_tweets'] = neut
traffic['Vad_comp_tweets'] = comp

traffic.head().T
````
